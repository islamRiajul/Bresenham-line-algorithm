```python
import pandas as pd
x=34
y=40
x1=46
y1=47
wd=0
dic={}
dx=(x1-x)
dy=(y1-y)
dy2=2*dy
dx2=2*dx
p=dy2-dx
k=0
while wd in range(dx):
    if p>=0:
        e=p
        x=x+1
        y=y+1
        a=x
        b=y
        p=p+dy2-dx2
        c=0
        keys = wd
        dic[keys] = e,a,b,c,p
    else:
        e=p
        x=x+1
        y=y
        a=x
        b=y
        p=p+dy2
        c=0
        keys = wd
        dic[keys] = e,a,b,p,c
    wd=wd+1
df = pd.DataFrame(dic.values(), columns=["P","X", "Y","Pk1","Pk2"]) 
df      

   
```
